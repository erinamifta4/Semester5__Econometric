## Chow Test
Chow Test pada dasarnya menguji apakah ada perubahan struktural dalam hubungan antar variabel di dua periode.

Jika menolak H0 (ada perbedaan), maka dilanjut ke Hausman Test.

Jika menerima H0 (tidak ada perbedaan), maka lanjut ke BP-LM Test

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy.stats import f
from linearmodels.panel import PanelOLS, RandomEffects

# **1. Persiapan Data dengan MultiIndex**
# Misalkan data Anda ada dalam `df` dengan kolom: 'province', 'year', 'grdp_employed', 'formal_sector', 'hdi', 'grdp_capita'
# Pastikan MultiIndex untuk panel data
df = data
df = df.set_index(['province', 'year']).sort_index()

# **2. Chow Test**
split_year = 2018

# Pisahkan data berdasarkan tahun tertentu
df1 = df[df.index.get_level_values('year') <= split_year]
df2 = df[df.index.get_level_values('year') > split_year]

# Tentukan variabel independen (X) dan dependen (Y)
X1 = df1[['grdp_employed', 'formal_sector', 'hdi']]
Y1 = df1['grdp_capita']
X2 = df2[['grdp_employed', 'formal_sector', 'hdi']]
Y2 = df2['grdp_capita']

# Tambahkan konstanta
X1 = sm.add_constant(X1)
X2 = sm.add_constant(X2)

# Model regresi untuk setiap subset
model1 = sm.OLS(Y1, X1).fit()
model2 = sm.OLS(Y2, X2).fit()

# Gabungkan data untuk model keseluruhan
X_full = sm.add_constant(df[['grdp_employed', 'formal_sector', 'hdi']])
Y_full = df['grdp_capita']
model_full = sm.OLS(Y_full, X_full).fit()

# Residual Sum of Squares (RSS)
RSS1 = model1.ssr
RSS2 = model2.ssr
RSS_full = model_full.ssr

# Jumlah observasi dan parameter
n1 = len(Y1)
n2 = len(Y2)
k = X1.shape[1]

# Statistik F untuk Chow Test
numerator = (RSS_full - (RSS1 + RSS2)) / k
denominator = (RSS1 + RSS2) / (n1 + n2 - 2 * k)
F_stat = numerator / denominator

# P-value
p_value = 1 - f.cdf(F_stat, k, n1 + n2 - 2 * k)

print("Chow Test:")
print("Nilai F-statistik:", F_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print(f"Hipotesis Alternatif (H1) (tolak H0): Terdapat perbedaan signifikan antara dua periode (tahun <= {split_year} dan tahun > {split_year})")
else:
    print(f"Hipotesis Nol (H0) (gagal tolak H0): Tidak terdapat perbedaan signifikan antara dua periode (tahun <= {split_year} dan tahun > {split_year})")
```

# Komparasi dgn Hausman Test

```python
# Uji Hausman

import pandas as pd
import numpy as np
from linearmodels.panel import RandomEffects, PanelOLS
import statsmodels.api as sm
from scipy import stats

# Pastikan data memiliki struktur panel (MultiIndex berdasarkan id dan waktu/tahun)
df_panel = data.set_index(['province', 'year'])  # 'id' harus sesuai dengan nama kolom ID data panel Anda

# Tentukan variabel dependen dan independen
Y = df_panel['grdp_capita']
X = df_panel[['grdp_employed', 'formal_sector', 'hdi']]
X = sm.add_constant(X)

# Estimasi model Random Effects
random_effects = RandomEffects(Y, X).fit()

# Estimasi model Fixed Effects dengan menggunakan PanelOLS dan mengaktifkan entity_effects
fixed_effects = PanelOLS(Y, X, entity_effects=True).fit()

# Hitung nilai statistik Hausman
b_random = random_effects.params
b_fixed = fixed_effects.params
cov_random = random_effects.cov
cov_fixed = fixed_effects.cov

# Selisih parameter dan matriks kovarians
b_diff = b_fixed - b_random
cov_diff = cov_fixed - cov_random

# Hitung nilai statistik Hausman
hausman_stat = b_diff.T @ np.linalg.inv(cov_diff) @ b_diff
df_hausman = b_diff.shape[0]
p_value = 1 - stats.chi2.cdf(hausman_stat, df_hausman)

print("Hausman Test Statistik:", hausman_stat)
print("P-value:", p_value)

# Interpretasi hasil uji Hausman
if p_value < 0.05:
    print("Hipotesis Alternatif (H1) / Tolak H0: Model Fixed Effects lebih sesuai daripada Random Effects")
else:
    print("Hipotesis Nol (H0) / Terima H0: Model Random Effects lebih sesuai daripada Fixed Effects")
```

### Mengapa Hausman Test Lebih Diprioritaskan?

Validitas Model: Komparasi manual (berdasarkan R-squared dan F-statistic) hanya menunjukkan kemampuan model menjelaskan variasi data. Namun, tidak memastikan asumsi RE terpenuhi.

Korelasi Efek Individual: Hausman Test menguji apakah efek individual berkorelasi dengan variabel independen, faktor kunci dalam memilih model. Jika ada korelasi, RE menjadi bias, dan FE harus digunakan.

Robustness: Hausman Test langsung membandingkan koefisien FE dan RE untuk memastikan model yang lebih sesuai.
